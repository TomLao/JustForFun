基于图的推荐系统

[详细解释](https://blog.csdn.net/Gamer_gyt/article/details/51694250)

```python
# G:二分图  alpha:随机游走概率  root:起始节点  max_stap:最大走动步数
def PersonalRank(G, alpha, root, max_step):
    rank = dict()
    rank = {x:0 for x in G.keys()}
    rank[root] = 1
    
    for k in range(max_step):
        tmp = {x:0 for x in G.keys()}
        
        #去节点i和他的出边尾节点集合ri，i是顶点，ri是与其相连接顶点和边的权重
        for i, ri in G.items():
            
            for j, wij in ri.items():
                
                tmp[j] += alpha * rank[i] / (1.0 * len(ri))
                
        tmp[root] += (1 - alpha)
        rank = tmp
        
        print('iter:' + str(k) + "\t", end="")
        for key, value in rank.items():
            print('%s:%.3f, \t' % (key, value), end="")
        print()
        
    return rank
```

```python
G = {'A' : {'a' : 1, 'c' : 1},
    'B' : {'a' : 1, 'b' : 1, 'c':1, 'd':1},  
     'C' : {'c' : 1, 'd' : 1},  
     'a' : {'A' : 1, 'B' : 1},  
     'b' : {'B' : 1},  
     'c' : {'A' : 1, 'B' : 1, 'C':1},  
     'd' : {'B' : 1, 'C' : 1}}
PersonalRank(G, 0.85, 'A', 100)
```
